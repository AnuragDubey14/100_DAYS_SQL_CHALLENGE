### Problem 1
Display the names of athletes who won a gold medal in the 2008 Olympics and whose height is greater than the average height of all athletes in the 2008 Olympics.

```python
%%sql
SELECT Name
FROM olympics
WHERE Year = 2008
    AND Medal = 'Gold'
    AND Height > (SELECT AVG(Height) FROM olympics WHERE Year = 2008);
```

### Problem 2
Display the names of athletes who won a medal in the sport of basketball in the 2016 Olympics and whose weight is less than the average weight of all athletes who won a medal in the 2016 Olympics.

```python
%%sql
select Name 
from olympics
    where year = 2016
    and sport='Basketball'
    and medal <> 'None' and 
    weight<(select avg(weight) from olympics where year = 2016 and medal <> 'None')
```

### Problem 3
Display the names of all athletes who have won a medal in the sport of swimming in both the 2008 and 2016 Olympics.

```python
%%sql
SELECT DISTINCT(Name)
FROM olympics
WHERE medal <> 'None' 
    AND sport = 'Swimming'
    AND Name IN (
        SELECT Name
        FROM olympics
        WHERE medal <> 'None' 
            AND sport = 'Swimming'
            AND year = 2008
    )
    AND Name IN (
        SELECT Name
        FROM olympics
        WHERE medal <> 'None' 
            AND sport = 'Swimming'
            AND year = 2016
    );
```

### Problem 4
Display the names of all countries that have won more than 50 medals in a single year.

```python
%%sql
SELECT COUNTRY, YEAR FROM OLYMPICS 
    WHERE MEDAL <> 'None' 
    GROUP BY COUNTRY,YEAR
    HAVING COUNT(*)>50
```

### Problem 5
Display the names of all athletes who have won medals in more than one sport in the same year.

```python
%%sql
SELECT Name, year, sport
FROM olympics
WHERE medal <> 'None' 
GROUP BY Name, year, sport
HAVING COUNT(*) > 1;
```

### Problem 6
What is the average weight difference between male and female athletes in the Olympics who have won a medal in the same event?

```python
%%sql
SELECT Event,
       AVG(CASE WHEN Sex = 'M' THEN Weight END) AS Male_Avg_Weight,
       AVG(CASE WHEN Sex = 'F' THEN Weight END) AS Female_Avg_Weight,
       AVG(CASE WHEN Sex = 'M' THEN Weight END) - AVG(CASE WHEN Sex = 'F' THEN Weight END) AS Avg_Weight_Difference
FROM olympics
WHERE Medal <> 'None'
GROUP BY Event
ORDER BY Avg_Weight_Difference DESC;
```
