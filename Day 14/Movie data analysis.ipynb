### Rank movies based on their ratings, showing the movie title and its rank and select top 10 movies.

```python
%%sql
SELECT *
FROM (
    SELECT title_x,
           DENSE_RANK() OVER (ORDER BY imdb_rating DESC) AS 'rating_rank'
    FROM movie
) AS ranked_movies
WHERE rating_rank <= 10;
```

### Create a pivot table that shows the average rating for each genre, with genres as columns and the average rating as values.

```python

%%sql
SELECT
    AVG(CASE WHEN genres LIKE '%Action%' THEN imdb_rating END) AS Action,
    AVG(CASE WHEN genres LIKE '%Adventure%' THEN imdb_rating END) AS Adventure,
    AVG(CASE WHEN genres LIKE '%Animation%' THEN imdb_rating END) AS Animation,
    AVG(CASE WHEN genres LIKE '%Biography%' THEN imdb_rating END) AS Biography,
    AVG(CASE WHEN genres LIKE '%Comedy%' THEN imdb_rating END) AS Comedy,
    AVG(CASE WHEN genres LIKE '%Crime%' THEN imdb_rating END) AS Crime,
    AVG(CASE WHEN genres LIKE '%Drama%' THEN imdb_rating END) AS Drama,
    AVG(CASE WHEN genres LIKE '%Family%' THEN imdb_rating END) AS Family,
    AVG(CASE WHEN genres LIKE '%Fantasy%' THEN imdb_rating END) AS Fantasy,
    AVG(CASE WHEN genres LIKE '%History%' THEN imdb_rating END) AS History,
    AVG(CASE WHEN genres LIKE '%Horror%' THEN imdb_rating END) AS Horror,
    AVG(CASE WHEN genres LIKE '%Music%' THEN imdb_rating END) AS Music,
    AVG(CASE WHEN genres LIKE '%Mystery%' THEN imdb_rating END) AS Mystery,
    AVG(CASE WHEN genres LIKE '%Romance%' THEN imdb_rating END) AS Romance,
    AVG(CASE WHEN genres LIKE '%Sci-Fi%' THEN imdb_rating END) AS SciFi,
    AVG(CASE WHEN genres LIKE '%Sport%' THEN imdb_rating END) AS Sport,
    AVG(CASE WHEN genres LIKE '%Thriller%' THEN imdb_rating END) AS Thriller,
    AVG(CASE WHEN genres LIKE '%War%' THEN imdb_rating END) AS War,
    AVG(CASE WHEN genres LIKE '%Western%' THEN imdb_rating END) AS Western
FROM movie;
```

### Use a Common Table Expression to find movies with ratings higher than the average rating.

```python
%%sql
WITH AvgRatingCTE AS (
    SELECT AVG(imdb_rating) AS avg_rating
    FROM movie
)

SELECT title_x, imdb_rating
FROM movie
CROSS JOIN AvgRatingCTE
WHERE imdb_rating > avg_rating;
```

### Calculate the total number of movies in each genre and the average rating, but only for genres with more than 5 movies

```python
%%sql
select genres, count(title_x) as 'number of movies', round(avg(imdb_rating),2) as average_rating
from movie
group by genres 
having ('number of movies')>5
order by 2 desc;
```

### Calculate the running total of ratings for each genre over the years, ordered by genre and release year.

```python
%%sql 
select genres,release_date,count(*) as movie_count
from movie
group by genres,release_date
order by 1,2
```
